```python
import pandas as pd
import requests
import json
import warnings
warnings.filterwarnings("ignore")
```

**3.0 Dados**

Obter e baixar dados

**3.1 Pontuações**

Obter pontuações dos jogadores nas rodadas

```python
#funcao que obtem dado de acordo com a rodada (input)
def GetPontuacao(rodada):
 url = 'https://api.cartola.globo.com/atletas/pontuados/'+str(rodada)
 resposta = requests.get(url)
 data = resposta.json()
 df = pd.DataFrame(data)
 #Normaliza a coluna "atletas" e cria um novo DataFrame
 df = pd.json_normalize(df['atletas'])
 #Remove o prefixo "scout." das colunas
 df.rename(columns=lambda x: x.replace('scout.', ''), inplace=True)
 df['rodada'] = rodada
 
 return df
```

```python
#primeira execucao
rodadas = 1
print("Finalizado rodada:",str(rodadas)+ "...")
df1= GetPontuacao(1)

#loop para gerar dados, até ultima rodada.
while rodadas != "Erro":
 try:
  rodadas = rodadas+1
  df= GetPontuacao(rodadas)
  df1 = df.append(df1)
  print("Finalizado rodada:",str(rodadas)+ "...")
  
 except:
  rodadas = "Erro"
  print("Fim")
```

```python
#visualizar dados gerados
display(df1.head(5))
```

```python
#salvar dado
df1.to_csv('bases/cartola_pontos.csv', index=False,mode='w') 
```

**3.2 Clubes**

Obter codigo do clube.

```python
url= 'https://api.cartola.globo.com/atletas/mercado'
resposta = requests.request("GET", url)
objetos = json.loads(resposta.text)
clubes = pd.json_normalize(objetos['clubes'].values())
```

```python
#ver dados
display(clubes.head(5))
```

```python
#salvar dado
clubes.to_csv('bases/cartola_clubes.csv', index=False,mode='w') 
```

**3.3 Posições**

Obter posiçoes dos jogadores

```python
posicoes = pd.json_normalize(objetos['posicoes'].values())
```

```python
#salvar dado
posicoes.to_csv('bases/cartola_posicoes.csv', index=False,mode='w') 
```

**3.3 Partidas**

Obter codigo de partidas

```python
#funcao que obtem dado de acordo com a rodada (input)
def GetPartidas(rodada):
 url = 'https://api.cartola.globo.com/partidas/'+str(rodada)
 resposta = requests.get(url)
 objetos = json.loads(resposta.text)
 df = pd.json_normalize(objetos['partidas'])
 df['rodada'] = rodada
 
 return df
```

```python
#primeira execucao
rodadas = 1
print("Finalizado rodada:",str(rodadas)+ "...")
df1= GetPartidas(1)

#loop para gerar dados, até ultima rodada.
while rodadas != "Erro":
 try:
  rodadas = rodadas+1
  df= GetPartidas(rodadas)
  df1 = df.append(df1)
  print("Finalizado rodada:",str(rodadas)+ "...")
  
 except:
  rodadas = "Erro"
  print("Fim")
```

```python
#visualizar dados gerados
display(df1.head(5))
```

```python
#salvar dado
df1.to_csv('bases/cartola_partidas.csv', index=False,mode='w') 
```
