```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

1.0 Bases

```python
#carregar dados
equipes_conta = (pd
.read_csv("C:/Users/eduar/OneDrive/Área de Trabalho/Python/cartola_fc/silver/bases/equipes_contra.csv")
)

equipes_pro = (pd
.read_csv("C:/Users/eduar/OneDrive/Área de Trabalho/Python/cartola_fc/silver/bases/equipes_pro.csv")
)

partidas_pro = (pd
.read_csv("C:/Users/eduar/OneDrive/Área de Trabalho/Python/cartola_fc/silver/bases/partidas_pro.csv")
)

partidas_contra = (pd
.read_csv("C:/Users/eduar/OneDrive/Área de Trabalho/Python/cartola_fc/silver/bases/partidas_contra.csv")
)

jogadores_pro = (pd
.read_csv("C:/Users/eduar/OneDrive/Área de Trabalho/Python/cartola_fc/silver/bases/jogadores_pro.csv")
)

setor_pro = (pd
.read_csv("C:/Users/eduar/OneDrive/Área de Trabalho/Python/cartola_fc/silver/bases/setor_equipes_pro.csv")
)

setor_contra = (pd
.read_csv("C:/Users/eduar/OneDrive/Área de Trabalho/Python/cartola_fc/silver/bases/setor_equipes_contra.csv")
)
```

2.0 Tratamento bases.

```python
#join clubes e setor_pro
dados_finais = (pd
.merge(jogadores_pro, setor_pro, 
       left_on=['equipe','posicao','rodada'],
         right_on=['equipe','posicao','rodada'], how='left')
)

#join clubes e setor_contra
dados_finais = (pd
.merge(dados_finais, setor_contra, 
       left_on=['adv','posicao','rodada'],
         right_on=['adv','posicao','rodada'], how='left')
)

#join equipes_pro
dados_finais = (pd
.merge(dados_finais, equipes_pro, 
       left_on=['equipe','rodada'],
         right_on=['equipe','rodada'], how='left')
)

#join equipes_contra
dados_finais = (pd
.merge(dados_finais, equipes_conta, 
       left_on=['adv','rodada'],
         right_on=['adv','rodada'], how='left')
)


#join partidas_pro
dados_finais = (pd
.merge(dados_finais, partidas_pro, 
       left_on=['equipe','rodada'],
         right_on=['equipe','rodada'], how='left')
)

#join partidas_contra
dados_finais = (pd
.merge(dados_finais, partidas_contra, 
       left_on=['equipe','rodada'],
         right_on=['equipe','rodada'], how='left')
)

# Encontre todas as colunas de texto
colunas_de_texto = dados_finais.select_dtypes(include=['object']).columns.tolist()

print("Colunas de Texto:")
print(colunas_de_texto)
```
